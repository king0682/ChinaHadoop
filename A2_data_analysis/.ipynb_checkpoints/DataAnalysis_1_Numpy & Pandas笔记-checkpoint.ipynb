### 广播特性/boardcasting
自动完成维度匹配的运算：简单来说，为了将就高维矩阵，将低维度扩展到高维（取最小公倍数），前提是：
1. 低维度矩阵的宽或高，等于某个高维度矩阵的宽或高

**注意：**后续深度学习搭建时，报错dimision not match时，就需要思考数据维度的变化，每一步做完之后处于什么样的形态。
![](./numpy-broadcast.png)

---

当操作两个array时，numpy会逐个dimension比较它们的shape，在下述情况下，两arrays会兼容和输出broadcasting结果：<br>

1. 相等
2. 其中一个为1，（进而可进行拷贝拓展与已知shape匹配）
3. 当两个ndarray的维度不完全相同的时候，rank较小的那个ndarray会被自动在前面加上一个一维维度，直到与另一个ndarray rank相同再检查是否匹配

比如求和的时候有，<font color='red'>**Result是B改造后的结果**</font>：
```python
Image (3d array):  256 x 256 x 3
Scale (1d array):              3
Result (3d array): 256 x 256 x 3

A      (4d array):  8 x 1 x 6 x 1
B      (3d array):      7 x 1 x 5
Result (4d array):  8 x 7 x 6 x 5

A      (2d array):  5 x 4
B      (1d array):      1
Result (2d array):  5 x 4

A      (2d array):  15 x 3 x 5
B      (1d array):  15 x 1 x 5
Result (2d array):  15 x 3 x 5
```

### 数据删除
DataFrame中的数据删除用drop完成  

**Note:  **
DataFrame的很多操作(包括drop)都有一个inplace参数，inplace=False则以返回值返回变换结果，inplace=True则在原变量直接变换


![](./axis.jpg)
<font color='red'>**关于axis的方向，请注意：**</font>  
* 如果是求运算的过程<font color='red'>**（动里面的瓤）**</font>，如：`mean()`,`add()`,`sum()`
    * axis=0是动态向下的过程，***sum(axis=0)为纵向求和***
    * axis=1是动态向右的过程，***sum(axis=1)为横向求和***
* 如果只是一个操作固有元素<font color='red'>**（动外面的框）**</font>，与numpy的操作不同，如`drop()`去掉某一行
    * axis=0，操作纵向堆积的index，横向排布，___所以drop(,axis=0)为删除行___
    * axis=1，操作横向堆积的column，纵向排布，___所以drop(,axis=1)为删除列___

````python
frame
````

````python
# 删除行：直接用index即可，堆积方向axis=0(默认为0)
frame.drop(["one", "two"], axis=0)

# 删除列：标出column的名称，堆积方向axis=1
frame.drop('debt', axis=1)
````

## Pandas--分组与聚合
目的：拿到数据后，对数据进行分组和聚合后得到统计值，帮助了解数据的分布，并予以相应的刻画。常见的三个函数：  

意义：
* 电商网站上，男性和女性的行为不同，需要将其提取字段进行分别分析
* 不同电商网站，不同年龄段的人行为模式也不同
* 电商网站上，不同地域的人行为模式也不同（江浙沪包邮）  

<font color='blue'>**依据以上的维度，将数据分成不同的Group，在每个组内做统计，对组内数据分布做刻画**</font>

* `info()`--字段、每个字段缺失值情况、每个字段的数据类型、数据占用资源(系统内存)大小  
* `.shape`--数据的维度和形状  
* `describe()`--数据的统计分布情况

常见操作流程：
* 先用groupby分组，再用sum,size,mean,median函数聚合


### 分组/groupby

````python
# 1.创建员工信息表

import pandas as pd
import numpy as np
# 一会儿绘图的时候，将图嵌入在notebook中，不要弹出来
%matplotlib inline
salaries = pd.DataFrame({
    'Name': ['BOSS', 'Jason', 'Jason', 'Han', 'BOSS', 'BOSS', 'Jason', 'BOSS'],
    'Year': [2016,2016,2016,2016,2017,2017,2017,2017],
    'Salary': [10000,2000,4000,5000,18000,25000,3000,4000],
    'Bonus': [3000,1000,1000,1200,4000,2300,500,1000]
})
salaries
````

````python
# 2.根据姓名进行分组
group_by_name = salaries.groupby('Name')
# group_by_name是pandas中groupby.DataFrameGroupBy的一个对象
group_by_name
````

### 聚合/[aggregate](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.aggregate.html)

````python
# 1.分组后加和
group_by_name.sum()

# 默认会按照index排序，也可以选择不排序
salaries.groupby('Name', sort=False).sum()

# 换一种写法
group_by_name.aggregate(sum)
````

````python
# 2.查看groupby对象groups属性（聚合条目在原DataFrame中的index）
print(group_by_name.groups)
print(len(group_by_name))
````

````python
# 3.多个变量group（先合并Name，再合并Year）
group_by_name_year = salaries.groupby(['Name', 'Year']).sum()
salaries.groupby(['Name', 'Year']).sum()

# 也可以用agg实现多column聚合
salaries.groupby(['Name', 'Year']).agg([np.sum, np.median, np.mean])

# 或者直接用agg+字符串给出分组信息
salaries.groupby(['Name', 'Year']).agg(['sum','median','mean'])
````

````python
# 4.分组并加和后，统计分组后的数据条数
group_by_name_year.size()
````

````python
# 5.均值
group_by_name.mean()
````

````python
# 6.中位数
group_by_name.median()
````

````python
# 7.统计信息
group_by_name.describe()
````

### 遍历分组
将每一组分组的分组折叠依据展开，以name分组，将具体信息一条条呈现

````python
for name,group in group_by_name:
    print(name)
    print("=====")
    print(group)
    print('\n')
````

````python
# 显示一个具体的group（每个人具体的详情）
group_by_name.get_group("Jason")
group_by_name.get_group("BOSS")
````

agg是aggregate的alias，可以替代使用

````python
# 对每个人的每个变量进行统计（sum,mean,std）
group_by_name.agg([np.sum, np.mean, np.std])
````

````python
# 对每一列可以采用不同的aggregate操作
group_by_name.agg({"Bonus": np.sum, "Salary": np.sum})

group_by_name.agg({"Bonus": np.sum, "Salary": np.sum, "Year": (lambda x: list(x)[0])})
````

````python
group_by_name.describe()
````

### [transform](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.transform.html) vs apply

作用：应用外部编写的函数，对每一个group的元素进行变换（强大之处：自定义想要做的变换）

````python
import pandas as pd
# index_col=0 指定第一列为index，parse_dates=[0]将第一列解析为日期类型（只接受：booleans, lists, and dictionaries）
nvda = pd.read_csv("./data/NVDA.csv", index_col=[0], parse_dates=[0])
nvda.head()

nvda.index
````

````python
# Note：用parse_dates解析后的日期型很重要，富含很多属性：星期几，对应一周第几天，工作日还是周末等信息
tmp = pd.read_csv('./data/NVDA.csv', parse_dates=['Date'])
date_col = tmp['Date'] # 一个中括号，取出来的是Series

# 查看一周的第几天
date_col.dt.dayofweek.head(5)

# 查看是星期几
date_col.dt.weekday_name.head(5)

# 将提取出来的信息加入DataFrame的新一列
tmp.loc[:,'Weekday Name'] = date_col.dt.weekday_name
tmp.head(5)
````

````python
# 求各个年份的均值：一定先用groupby分组，再用统计函数进行聚合
nvda.groupby(nvda.index.year).mean().head()

# 定义标准化函数
zscore = lambda x:(x-x.mean())/x.std()

# 转化函数1:transform()，调用标准化函数，对groupby中的每一组分类进行转化
transformed = nvda.groupby(nvda.index.year).transform(zscore)
transformed.head()

# 转化函数2:apply()，与transform效果相同
transformed = nvda.groupby(nvda.index.year).apply(zscore)
transformed.head()
````

````python
# 需求：对原数据close列，如果数值大于1.7则返回平方值，否则返回原值
def my_trans(x):
    return x**2 if x>1.7 else x

# 新增一列
nvda.loc[:,'new'] = nvda['Close'].apply(my_trans)
nvda.head(3)
````

````python
# 通过画图方式将变量进行刻画：原数据中的Adj Close & 标准化后的Adj Close
compare = pd.DataFrame({"Original Adj Close": nvda["Adj Close"], 
                        "Transormed Adj Close": transformed["Adj Close"]})
compare.plot()
````

````python
# 分别找出历史年份各个参数的最大值
nvda.groupby(nvda.index.year).transform("max").head()
````

### [filter](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.filter.html)

比如我们想要找出符合统计特征的一些内容，就像SQL里面的having, where语句

通俗而言，分组之后，想在组内加统计条件进行判断

````python
# 案例1:想找出组内加和大于4的分组（1只有两个，加和为2，不符合条件；而2有3个，加和为6，符合条件）
s = pd.Series([1,1,2,2,2,3,4,4,5])
s
s.groupby(s).groups
s.groupby(s).filter(lambda x: x.sum() > 4)
````

````python
# 案例2：找出分组后的B列中，每组元素个数大于2的item
df = pd.DataFrame({"A": np.arange(8), "B":list("aaabbbcc")})
df
df.groupby("B").filter(lambda x: len(x) > 2)
````

````python
# 案例3:找出nvda股票月均价>50和日均价>100的数据item
nvda.head()
nvda.groupby([nvda.index.year, nvda.index.month]).filter(lambda x: x["Adj Close"].mean() > 50).head()
nvda.groupby([nvda.index.year, nvda.index.month,nvda.index.day]).filter(lambda x:  x["Adj Close"]>100).head()
````

### 小结 
Group by: split-apply-combine
- 首先第一步是分离数据split，按照一定的规则把数据分成几类。
- 第二步是对每一部分数据都做一定的操作，这个操作可以是汇总操作aggregate，可以是一个变换transform，也可以是过滤数据filter。<font color='red'>**transform, apply, filter很重要**</font>
    * apply：用在整个dataframe上
    * transform：做完分组后，对分组做某种形式对变换
    * filter：分完组后，基于对这个组统计的结果做判断
- 最后一步就是把处理过的数据再合成一张DataFrame。

## Pandas--多表合并与拼接

目的：手头的几份数据并没有存储在一张表中，对几份数据通过关联字段进行合并

- concat
- append

````python
# 准备几张DataFrame准备拼接
import pandas as pd
import numpy as np
df1 = pd.DataFrame({'apts': [55000, 60000],
                   'cars': [200000, 300000],},
                  index = ['Shanghai', 'Beijing']
                  )
df1

df2 = pd.DataFrame({'cars': [150000, 120000],
                    'apts': [25000, 20000],
                   },
                  index = ['Hangzhou', 'Nanjing'])
df2

df3 = pd.DataFrame({'apts': [30000, 10000],
                   'cars': [180000, 100000],'mar': [180000, 100000]},
                  index = ['Guangzhou', 'Chongqing'])
df3
````

### [concat](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.concat.html)

物理形状：在行或者列方向插入item

#### column对齐，一行行拼接插入item（横向插入，纵向对齐）

````python
# 将上面的df1,df2,df3进行拼接
frames = [df1, df2, df3]
result = pd.concat(frames)
result
````

````python
# 给加入的每一个df命一个名字keys
result2 = pd.concat(frames, keys=['x','y','z'])
result2

# 根据keys和column取items
result2.loc["y","apts"]
````

#### 纵向插入，与每行的index对齐

````python
df4 = pd.DataFrame({'salaries': [10000, 30000, 30000, 20000, 15000]},
                  index = ['Suzhou', 'Beijing', 'Shanghai', 'Guangzhou', 'Tianjin'])
df4
````

````python
# axis=1：横向堆积，纵向排布
result3 = pd.concat([result, df4], axis=1)
result3

# 用inner去掉拼接两者一方有NaN的item(result或df4为空的部分)
result3 = pd.concat([result, df4], axis=1, join='inner')
print(result3)
````

### Merge(Join)

两张表或者两个DataFrame基于某一个或者某几个字段做连接的函数

````python
df1 = pd.DataFrame({'apts': [55000, 60000, 58000],
                   'cars': [200000, 300000,250000],
                  'city': ['Shanghai', 'Beijing','Shenzhen']})
df1

df4 = pd.DataFrame({'salaries': [10000, 30000, 30000, 20000, 15000],
                  'city': ['Suzhou', 'Beijing', 'Shanghai', 'Guangzhou', 'Tianjin']})
df4
````

````python
# 即使不指定字段，也会寻找相同字段连接
result = pd.merge(df1, df4, on='city')
result

result = pd.merge(df1, df4)
result
````

````python
# 外连接(两张表任何一张表有就会填入数值，没有则用NaN占空)
result = pd.merge(df1, df4, on='city', how='outer')
result

# 内连接（两表中一方全NaN的，就删去不要；而outer则是只要一方有东西，就留下来，除非两方都是NaN）

# 右关联（以右表为主，右表有的全都要）
result = pd.merge(df1, df4, on='city', how='right')
result

# 左关联（以左表为主，左表有的全都要）
result = pd.merge(df1, df4, on='city', how='left')
result
````

````python
import pandas as pd
# 1.导入评分表格ratings_small.csv
ratings = pd.read_csv('./data/ratings_small.csv')
ratings.head()
````

````python
# 2.根据movieId进行分组（统计每部电影有多少人打分）
# 形式一：通过agg('count')聚合
ratings.shape
cout = ratings.groupby("movieId").agg({"timestamp":"count"})
cout.head()

# 形式二：通过size()聚合
ratings.groupby("movieId").size().head()

# 形式三：在计算每部电影打分人数的基础上，计算每部电影的平均分
avg_score = ratings.groupby("movieId").agg({"rating":"mean","userId":"count"}).rename(columns={"rating":"averRate","userId": "userNum"})
avg_score.head()
avg_score.shape
# ratings.info()
````

表格2：电影的具体信息  movies_metadata.csv  
练习内容：
    * 取出某些字段
    * 处理有问题数据：用try...except...来处理
    * 数据关联：pd.merge()

````python
# 1.读入信息
meta = pd.read_csv('./data/movies_metadata.csv')
meta.columns
meta.head()
````

````python
# 2.找出电影的名字和id
# 正规写法
meta.loc[:,["id","title"]].head()

# 简写也可以实现（不推荐）
meta[["id","title"]].head()
````

````python
# 3.清洗脏数据：
# 任务：id字段本应为int，如果读入str就会报错。为了避免出现问题，就会将字符串转成int型
movie_titles = meta.loc[:,["id","title"]]
for i in range(movie_titles.shape[0]):
    try:
        movie_titles.loc[i, "id"] = int(movie_titles.loc[i, "id"])
    except ValueError:
        movie_titles.loc[i, "id"] = 999999999

# 当然也可以用pandas来转换成int
# movie_titles.loc[:,"id"] = pd.to_numeric(movie_titles.loc[:,"id"])
````

````python
# 4.如果想做数据关联，就用pd.merge
# 任务：将电影均分avg_score与电影名称movie_titles表格拼接起来
avg_score.head() # 默认将movieId作为index
movie_titles.head() # 
````

````python
# 内连接方式，left_index=True：左表将index作为关联字段，right_on="id"右表将id作为关联字段）
movie_name_ratings = pd.merge(avg_score, movie_titles, how="inner", left_index=True, right_on="id")
movie_name_ratings.head()
movie_name_ratings.shape

# 对上述关联结果以userNum和averRate两个字段做降序排序（先按照userNum排序，如果userNum相同时再按照averRate进行排序）
movie_name_ratings.sort_values(by=["userNum", "averRate"], ascending=False).head(5)
````

### 项目2：处理缺失数据
* 任务内容：处理脏数据（空数据NaN）
* 背景内容：处理Montreal共享单车数据bikes.csv（拉丁文），共7条路线，每条路线记录每天有多少人经过。

#### 做法1：删除缺失值

````python
# 再命令行中概览bikes.csv的大致情况，发现是乱码
!head -5 ./data/bikes.csv

# 下面使用正常的解码打开会报错：UnicodeDecodeError: 'utf-8' codec can't decode byte 0xe9
# bikes = pd.read_csv("./data/bikes.csv")

# 修改之后，使用latin-1解码，但是分隔符不对，需要指定;为分隔符，声明sep
bikes = pd.read_csv("./data/bikes.csv", encoding="latin-1")
bikes.head()

# 指定sep分隔符，将Date列解析为Date型，index字段为Date，dayfirst=True为"年-月-日"格式，而dayfirst=False为"年-日-月"格式
bikes = pd.read_csv('./data/bikes.csv', sep=';', parse_dates=['Date'], \
                    encoding='latin-1', dayfirst=True, index_col='Date')
bikes.head()
````

````python
# 查看数据信息
bikes.shape
bikes.info()
bikes.describe()
````

````python
# 对缺失数据进行处理：第二列和最后一列都是NaN，使用dropna()默认删除行
# 1.只要一行中有一个有缺失值，就删掉整行值
bikes.dropna()

# 2.使用how="all"进行表示：当一行中所有值均为NaN时，将这一行删除
bikes.dropna(how="all").head()
bikes.dropna(how="all").shape

# 3.若想删除某一列的话，则指定axis=1，即可将某个全空字段删除
bikes.dropna(how="all", axis=1).head()
bikes.dropna(how="all", axis=1).shape
````

#### 做法2：填充缺失值

````python
# 1.算出每个日期的平均值(sum(有效数据)/有效item数)：横向排布的数据
bikes.mean(1).shape

# 2.将第一步算出的数值填入NaN数据处，只能将单个缺失值的item的index逐个列为column，一列填写同一个缺失值
bikes.T.fillna(bikes.mean(1)).T.head()
````

````python
# 任务2：取出Berri 1字段，并将index日期的星期作为新一列标在最后。
# 1.取出某字段值作为DataFrame
berri_bikes = bikes[["Berri 1"]]
berri_bikes.head()

# 2.取出DataFrame索引，并转化成一周的第几天模式
berri_bikes.loc[:,"weekday"] = berri_bikes.index.weekday
berri_bikes.head()

# 3.按照星期做分组后用mean进行聚合，并将index进行英文化
weekday_counts = berri_bikes.groupby("weekday").mean()
weekday_counts.index = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekday_counts

# 4.将日期聚合后的数据用matplotlib画图
%matplotlib inline
weekday_counts.plot(kind="bar")
````

````python
# 统计7条路线的平均值，按照日期绘制均值折线图
# 1.取出bikes的平均值，转化成DataFrame，并将该列命名column为"bikes_avg"
bikes_avg = bikes.mean(axis=1).to_frame()
bikes_avg.columns = ["bikes_avg"]
bikes_avg.head()

# 2.取出日期进行分组
bikes_avg.loc[:,"weekday"] = bikes_avg.index.weekday
bikes_avg.head()

# 3.根据日期分组后用mean聚合，组成新的名为weekday_avg的DF
weekday_avg = bikes_avg.groupby("weekday").mean()
weekday_avg.index = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekday_avg

# 4.将每天的车流总量进行画图绘制
bikes_avg["bikes_avg"].plot()

# 5.画出每周平均的柱状图
weekday_avg.plot(kind="bar")
````

### 哪个地铁口附近的房子最多

````python
# 注意是地铁口，而不是哪条线，九龙山站在14号线和7号线
# 需要提取并且用这个信息，就要用正则表达式提取出来，而只判断是否集中供热，就直接判断某个字符串是否在所属字段中即可。
# lj_data.head(3)
import re
def get_subway_info(string, pattern, n):
    result = re.search(pattern, string)
    if result:
    # 这里group表示整个搜索的string，而1,2,3,4表示具体提取的某个括号
        return result.group(n)
    else:
        return ''
````

````python
pattern = '距离(.+线(\(.*段\))?)(.+站)(\d{1,})米'
# 对某列字段进行同一操作，使用apply()函数
lj_data.loc[:,"subway_station"] = lj_data['extra_info'].apply(lambda x:get_subway_info(x, pattern, 3))
lj_data["subway_station"].value_counts()

# 从下面可见，不在地铁站为1086，其余以此向下
````

### 地铁附近的房子平均价格 比 非地铁的高多少

### 地铁附近的房源离地铁平均距离

````python
pattern = '(\d+)米'
lj_data.loc[:,'subway_distance'] = lj_data['extra_info'].apply(lambda x:get_subway_info(x, pattern,1))

# 直接初mean会出现inf的错误，所以先要确保横向值为True，否则会出现‘’无法转换成int的情况
lj_data.loc[lj_data.near_subway==True,'subway_distance'].astype(int).mean()
````
